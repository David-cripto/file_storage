```python
import matplotlib.pyplot as plt
from pathlib import Path
```

```python
path_to_eval = "/cache/david/dsbm_code/DSBM_AdvBM/evaluation/"
exp_name = "exp_name=celeba_male2female_dsbm_params_minibatch_ot_eps_1,first_coupling=ref,first_num_iter=20000,gamma_max=0.01,gamma_min=0.01,method=dbdsb,num_iter=20000,num_steps=100,path_to_save_info=/"
path_to_eval = Path(path_to_eval)
f_b_ar = ["f"]
```

```python
res_dict = {fb: {} for fb in f_b_ar}
imf_iters_dict = {fb: {"imf_iter":[]} for fb in f_b_ar}
for fb in f_b_ar:
    for imf_iter_res in (path_to_eval / exp_name / fb).glob("*"):
        imf_iter = int(imf_iter_res.name)
        imf_iters_dict[fb]["imf_iter"].append(imf_iter)

imf_iters_dict = {fb: {"imf_iter":sorted(imf_iters_dict[fb]["imf_iter"])} for fb in f_b_ar}
for fb in f_b_ar:
    for imf_iter in imf_iters_dict[fb]["imf_iter"]:
        imf_iter_res = path_to_eval / exp_name / fb / str(imf_iter)
        for imf_dir in imf_iter_res.glob("*"):
            if imf_dir.is_file():
                continue
            for metric_txt in imf_dir.glob("*.txt"):
                with open(metric_txt, "r") as f:
                    line = f.readlines()[0]
                    value = float(line.split(" ")[-1])
                name_metric = metric_txt.name.split("_")[0]
                if name_metric in res_dict[fb]:
                    res_dict[fb][name_metric].append(value)
                else:
                    res_dict[fb][name_metric] = [value]
```

```python
for fb in f_b_ar:
    for metric in res_dict[fb]:
        plt.figure()
        plt.plot(imf_iters_dict[fb]['imf_iter'], res_dict[fb][metric], label = metric)
        plt.xlabel("IMF iter")
        plt.legend();
```
